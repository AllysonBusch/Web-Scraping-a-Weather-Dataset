```python
import requests
import pandas as pd

from bs4 import BeautifulSoup
```

```python
# Parameterizing my location: Orlando, FL
latitude = 28.5383396
longitude = -81.3792418
```

```python
# Requesting the weather data for Orlando, FL
url = "http://forecast.weather.gov/MapClick.php?lat="+str(latitude)+"&lon="+str(longitude)
page = requests.get(url)
```

```python
# Parsing the page content and storing in a beautiful soup object
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
# Find tag to grab the next seven days forecast data
find_day = soup.find_all('div')
find_day
```

```python
seven_day = soup.find(id='seven-day-forecast')
seven_day
```

```python
# After searching for the tag and the container
forecast_items = seven_day.find_all(class_="tombstone-container")

# Checking the data with tonight's forecast
tonight = forecast_items[0]
print(tonight.prettify())
```

```python
# storing the periods
period_tags = seven_day.select(".tombstone-container .period-name")
periods = [pt.get_text() for pt in period_tags]
```

```python
# Storing the summary forecasts
summary_descs = [sd.get_text() for sd in seven_day.select(".tombstone-container .short-desc")]
```

```python
# Storing the temperatures
temperatures = [t.get_text() for t in seven_day.select(".tombstone-container .temp")]
```

```python
# Storing the detail forecasts
descriptions = [d["title"] for d in seven_day.select(".tombstone-container img")]
```

```python
print(periods)
print(summary_descs)
print(temperatures)
print(descriptions)
```

```python
# Creating a pandas dataframe to store the seven days of forecasting
weather = pd.DataFrame({
    "Period": periods,
    "Summary Forecast": summary_descs,
    "Temperature": temperatures,
    "Detail Forecast": descriptions
})
weather
```

```python
# Converting the temperature to C and displaying both
# This is extracting just the numeric value
temp_nums = weather["Temperature"].str.extract("(?P<temp_num>\d+)", expand = False)
weather["temp_num"] = temp_nums.astype('int')
weather
```

```python
# Converting this to C and storing it
temp_cent = pd.to_numeric(temp_nums)
temp_cent = (temp_cent-32)*(5/9)
temp_cent
```

```python
# Checking the average temperature
weather["temp_num"].mean()
```

```python
# storing the C value in the dataframe
weather["temp_C"] = temp_cent.astype('int').astype('str')
weather
```

```python
# Flagging night time
is_night = weather["Period"].str.contains("night|Night")
weather["Is_Night"] = is_night
weather
```

```python
# Preparing the final dataset and getting rid of redundant columns
weather = weather.drop(columns=["temp_num"])
weather
```
